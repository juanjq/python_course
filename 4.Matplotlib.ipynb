# <font color='gold'>**Fillbetween**</font>

```python
import scipy.stats as ss
x=np.linspace(0,3,200)
y1=ss.norm.pdf(x,1,0.3)
y2=np.sin(x)
plt.plot(x,y1,'-k')
plt.plot(x,y2,'--r',markersize=2.6)

plt.fill_between(x,y1,0.3,                      color='c')
plt.fill_between(x,y1,y2, where=y2<y1,          color='g')
plt.fill_between(x,y2,0,  where=abs(x-2.3)<0.2, color='orange')

plt.show()
```

# <font color='gold'>**Polynomial fits**</font>

First to import the classes:

```python
import httpimport
with httpimport.remote_repo(['regression_plot'], 'https://gist.githubusercontent.com/juanjq/da4e84d0afe19063ef24dacffc944dff/raw/19673a94a7d59798923326be8fe3181e3a44bd06'):
    import regression_plot as reg
```

Also we generate some data to do prove

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.random.normal(15,20,100)
y=list(map(lambda x: abs(x*np.random.normal(6,2,1)),x))
```

* **Linear regression**

```python
linreg=reg.Linear(x,y)

#print results
linreg.results()

#get values 
A,uA=linreg.A,linreg.uA
B,uB=linreg.B,linreg.uB
R=linreg.R


fig, ax = plt.subplots()
plt.plot(x,y,'.r')

#plot the line
linreg.plot(ax,color='darkblue',linestyle='-.')

plt.show()
```

* **Polynomial regression**

```python
polreg=reg.Polynomial(x,y,degree=3)

#print results
polreg.results()

#get coefficients 
coeficients=polreg.coefs()
print(coeficients)


fig, ax = plt.subplots()
plt.plot(x,y,'.r')

#plot the polynomial
polreg.plot(ax,color='darkblue',linestyle='-.')

plt.show()
```

# <font color='gold'>**Plot an ellipse**</font>

To plot an ellipse we have to import and give the parameters

```python
import matplotlib.pyplot as plt

Xcenter,Ycenter=0.5,1
A,a=1,0.5
degrees=44




from matplotlib.patches import Ellipse

fig, ax=plt.subplots()

ellipse = Ellipse(xy=(Xcenter,Ycenter),width=A,height=a,
                      angle=degrees, 
                      edgecolor='k', fc='None', lw=2)

ax.add_patch(ellipse)
ax.autoscale()




plt.show()
```
